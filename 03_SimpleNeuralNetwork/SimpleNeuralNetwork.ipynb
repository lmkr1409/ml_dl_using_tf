```python
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelBinarizer
```

```python
X, y = datasets.load_iris(return_X_y=True)
```

```python
lbr = LabelBinarizer()
y_label = lbr.fit_transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_label, test_size=0.2, random_state=35)
```

```python
def create_model(input_features, output_labels):
    model = Sequential()
    model.add(Dense(16, input_shape=[input_features,], activation = tf.nn.relu))
    model.add(Dense(8, activation = tf.nn.relu))
    model.add(Dense(output_labels, activation = tf.nn.softmax))
    model.compile(
        loss = tf.keras.losses.CategoricalCrossentropy(from_logits=True),
        optimizer = 'Adam',
        metrics = ['accuracy'])
    model.summary()
    return model
model = create_model(4,3)
```

```python
tb = tf.keras.callbacks.TensorBoard('logs/sample/keras')
es = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=3)
model.fit(X_train, y_train, epochs=500, callbacks = [tb, es], validation_split = 0.1)
```

```python
model.evaluate(X_test, y_test)
```
