```python
import tensorflow as tf
import pandas as pd
import numpy as np
import os
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelBinarizer
```

```python
DATA_SET_PATH = ''
df = pd.read_csv(os.path.join(DATA_SET_PATH, 'Iris.csv'))
```

```python
df = df.drop(columns=['Id'])
df.head()
```

```python
X = df.drop(['Species'], axis = 1).values
y = df['Species'].values
```

```python
tag_encoder = LabelBinarizer()
y_label = tag_encoder.fit_transform(y)
```

```python
y_label[0]
```

```python
tag_encoder.classes_
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_label, test_size=0.2, random_state=35)
```

```python
def create_model(num_of_features, output_lables):
    model = Sequential()
    model.add(Dense(output_lables, input_shape=(num_of_features,), activation = 'softmax'))
    model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
    model.summary()
    return model
model = create_model(len(X_train[0]), len(y_train[0]))
```

```python
callback = tf.keras.callbacks.EarlyStopping(monitor='loss', patience=3)
history = model.fit(X_train, y_train, epochs=1000, callbacks=[callback])
```

```python
model.evaluate(X_test, y_test)
```
